```python
import tensorflow as tf

from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
datagen_foot = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data_foot = datagen_foot.flow_from_directory("Images/Foot",class_mode='binary',
                                         color_mode='grayscale',
                                         target_size=(256, 145),
                                         subset='training')
test_data_foot = datagen_foot.flow_from_directory("Images//Foot",class_mode='binary',
                                         color_mode='grayscale',
                                         target_size=(256, 145),
                                         subset='validation')
```

```python
datagen_ankle = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_data_ankle = datagen_ankle.flow_from_directory("Images/Ankle",class_mode='binary',
                                         color_mode='grayscale',
                                         target_size=(256, 256),
                                         subset='training')
test_data_ankle = datagen_ankle.flow_from_directory("Images/Ankle",class_mode='binary',
                                         color_mode='grayscale',
                                         target_size=(256, 256),
                                         subset='validation')
```

```python
plt.imshow(test_data_ankle[0][0][0])
```

```python
plt.imshow(test_data_foot[0][0][20])
```

```python
def build_model(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))

    model.add(layers.Flatten())
    model.add(layers.Dense(16,activation='relu'))
    model.add(layers.Dense(1,activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

```python
foot_state_classifier = build_model()
ankle_state_classifier = build_model()
```

```python
foot_state_classifier.summary()
```

```python
ankle_state_classifier.summary()
```

```python
foot_state_classifier.fit(train_data_foot, epochs=20)
```

```python
ankle_state_classifier.fit(train_data_ankle, epochs=2)
```

```python
foot_score = foot_state_classifier.evaluate(test_data_foot)
ankle_score = foot_state_classifier.evaluate(test_data_ankle)
print("\n-----------------------------------------------------")
print(f'Foot Score, Loss: {foot_score[0]}, Accuracy: {foot_score[1]}')
print(f'Ankle Score, Loss: {ankle_score[0]}, Accuracy: {ankle_score[1]}')
```
